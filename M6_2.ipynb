```python
from sklearn.cluster import KMeans
```

```python
import pandas
```

```python
data=pandas.read_excel('dummy_data_for_kmeans_discussion.xlsx')
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data[['X']],data[['Y']])
```

```python
kmeans = KMeans(n_clusters=3).fit(data[['X','Y']])
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
import numpy as np
plt.scatter(data[['X']],data[['Y']], c=np.array(kmeans.labels_))
```
